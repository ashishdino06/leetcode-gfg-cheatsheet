```python
#unique email addresses
emails = ["test.email+alex@leetcode.com","test.e.mail+bob.cathy@leetcode.com","testemail+david@lee.tcode.com"]
domains=[email.split('@')[1] for email in emails]
local=[email.split('@')[0] for email in emails]

for i in range(len(local)):
  local[i] = local[i].replace('.','').partition('+')[0]


result=[u+'@'+d for u,d in zip(local,domains)]
print(local)
print(result)
print(len(set(result)))
# print(domains)
```

```python
#regex appraoch
import re
def numUniqueEmails(emails):
    expression, emails_set = r'([a-z0-9.]+)\+?.*@(\S+)', set()

    for email in emails:
        matches = re.findall(expression, email)
        print(matches)
        email = re.sub(r'\.', '', matches[0][0]) + '@' + matches[0][1]
        print(email)
        emails_set.add(email)
    print(len(emails_set))
emails = ["test.email+alex@leetcode.com","test.e.mail+bob.cathy@leetcode.com","testemail+david@lee.tcode.com"]

numUniqueEmails(emails)
```

```python
#duplicatezeros
arr = [1,0,2,3,0,4,5,0]
print(arr)
i =0
n = len(arr)-1

while i <=n:
  if arr[i]==0:
    for j in range(n,i,-1):
      arr[j]=arr[j-1]
    i+=2
  else:
    i+=1
print(arr)
```
