```python
#peak element
class Solution:
    def peakElement(self, arr):
        if len(arr)==1:
            return 0
        for i in range(len(arr)):
            if i==0:
                if arr[i]>arr[i+1]:
                    return i
            elif i==len(arr)-1:
                if arr[i]>arr[len(arr)-2]:
                    return i
            else:
                if arr[i]>arr[i+1] and arr[i]> arr[i+1]:
                    return i
```

```python
class Solution:
    def getPairs(self, arr):
        seen=set()
        res = set()
        for ele in arr:
            if -ele in seen:
                res.add((min(-ele,ele),max(ele,-ele)))
            seen.add(ele)
        return sorted(res)



```
